# 📈 Stock Price Prediction using RNN, LSTM and GRU

## 1️⃣ Datasetni yuklash va dastlabki ko‘rib chiqish

Bu bo‘limda loyiha uchun kerakli kutubxonalar chaqiriladi va JCI_data.csv dataset yuklanadi. Dataset pandas yordamida DataFrame formatiga o‘tkazilib, uning dastlabki qatorlari (head()) va o‘lchami (shape) tekshiriladi.

Bu bosqich dataset strukturasi, ustunlar soni va ma’lumotlar hajmini tushunishga yordam beradi.

```python
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import MinMaxScaler
import torch.nn as nn
```

```python
df = pd.read_csv("JCI_data.csv")
df.head()
```

```python
df.shape
```

## 2️⃣ Data preprocessing va normalization

Modelni samarali ishlashi uchun datasetdan keraksiz ustunlar (date, Name) olib tashlanadi. Qolgan numeric ustunlar MinMaxScaler yordamida 0 va 1 oralig‘iga normallashtiriladi.

Normalization deep learning modellarining tezroq va barqaror o‘rganishiga yordam beradi.

```python
data = df.drop(columns=["date","Name"]).values
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

## 3️⃣ Time-series sequence yaratish

Bu bosqichda time-series forecasting uchun sliding window usuli orqali ketma-ketliklar (sequence) yaratiladi.

Har bir sequence uzunligi 30 timestep qilib tanlanadi. Model oldingi 30 ta qiymat asosida keyingi qiymatni bashorat qilishni o‘rganadi.

Bu usul RNN, LSTM va GRU kabi sequential modellar uchun juda muhim hisoblanadi.

```python
def seq(data,seq_len):
    x,y = [],[]
    for i in range(len(data)-seq_len):
        x.append(data[i:i+seq_len])
        y.append(data[i+seq_len])
    return np.array(x),np.array(y)
x,y = seq(data,30)
```

## 4️⃣ Train va Test datasetlarga ajratish

Yaratilgan sequence dataset 80% training va 20% testing qismlarga bo‘linadi.

So‘ngra numpy arraylar PyTorch tensorlariga o‘tkazilib DataLoader orqali batchlarga ajratiladi.

Bu modelni samarali train qilish va generalization sifatini tekshirish imkonini beradi.

```python
tr_size = (len(data) * 80) // 100
x_tr = x[:tr_size]
y_tr = y[:tr_size]
x_te = x[tr_size:]
y_te = y[tr_size:]
x_tr = torch.tensor(x_tr,dtype=torch.float32)
x_te = torch.tensor(x_te,dtype=torch.float32)
y_tr = torch.tensor(y_tr,dtype=torch.float32)
y_te = torch.tensor(y_te,dtype=torch.float32)
data_tr = torch.utils.data.TensorDataset(x_tr,y_tr)
data_te = torch.utils.data.TensorDataset(x_te,y_te)
tr_load = torch.utils.data.DataLoader(data_tr,batch_size=128)
te_load = torch.utils.data.DataLoader(data_te,batch_size=128)
```

## 5️⃣ RNN, LSTM va GRU modellarini yaratish va train qilish

Bu bo‘limda uch xil deep learning modeli quriladi:

RNN
LSTM
GRU

Har bir model sequential ma’lumotlarni o‘rganish uchun ishlatiladi.

Model train qilish jarayonida:

optimizer → Adam
loss function → MSELoss

tanlanadi va epochlar davomida model o‘qitiladi.

Bu bosqichda uchta modelning ishlash samaradorligi taqqoslanadi.

```python

class RNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.rnn = nn.RNN(
            input_size=5,
            hidden_size=64,
            batch_first=True
        )
        self.fc = nn.Linear(64,5)
    def forward(self,a):
        out,_ = self.rnn(a)
        out = out[:,-1,:]
        out = self.fc(out)
        return out
        
class LSTM(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=5,
            hidden_size=64,
            batch_first=True
        )
        self.fc = nn.Linear(64,5)
    def forward(self,a):
        out,_ = self.lstm(a)
        out = out[:,-1,:]
        out = self.fc(out)
        return out
        
class GRU(nn.Module):
    def __init__(self):
        super().__init__()
        self.gru = nn.GRU(
            input_size=5,
            hidden_size=64,
            batch_first=True
        )
        self.lin = nn.Linear(64,5)
    def forward(self,a):
        res,_ = self.gru(a)
        res = res[:,-1,:]
        res = self.lin(res)
        return res
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model_RNN = RNN().to(device)
model_LSTM = LSTM().to(device)
model_GRU = GRU().to(device)
optim1 = torch.optim.Adam(model_RNN.parameters(),lr=0.001)
optim2 = torch.optim.Adam(model_LSTM.parameters(),lr=0.001)
optim3 = torch.optim.Adam(model_GRU.parameters(),lr=0.001)
crit = nn.MSELoss()    
```

```python
for epoch in range(10):
    model_RNN.train()
    model_LSTM.train()
    model_GRU.train()
    for x_batch , y_batch in tr_load:
        x_batch = x_batch.to(device)
        y_batch = y_batch.to(device)
        optim1.zero_grad()
        optim2.zero_grad()
        optim3.zero_grad()
        pred1 = model_RNN(x_batch)
        pred2 = model_LSTM(x_batch)
        pred3 = model_GRU(x_batch)
        loss1 = crit(pred1,y_batch)
        loss2 = crit(pred2,y_batch)
        loss3 = crit(pred3,y_batch)
        loss1.backward()
        loss2.backward()
        loss3.backward()
        optim1.step()
        optim2.step()
        optim3.step()
    print(f"| Epoch:{(epoch+1):02d} | RNN_Loss:{loss1.item():.4f} | LSTM_Loss:{loss2.item():.4f} | GRU_Loss:{loss3.item():.4f} |")
```

## 6️⃣ Model natijalarini baholash va visualization

Train qilingan modellar test datasetda sinovdan o‘tkaziladi va bashorat qilingan qiymatlar original qiymatlarga qayta scale qilinadi (inverse_transform).

So‘ngra matplotlib yordamida:

real qiymatlar
RNN prediction
LSTM prediction
GRU prediction

grafiklarda taqqoslab ko‘rsatiladi.

Visualization modellar aniqligini solishtirish va qaysi model yaxshiroq ishlaganini aniqlash imkonini beradi.

```python
model_RNN.eval()
model_LSTM.eval()
model_GRU.eval()
m1,m2,m3=[],[],[]
with torch.no_grad():
    for xb,yb in te_load:
        xb = xb.to(device)
        yb = yb.to(device)
        res1 = model_RNN(xb)
        m1.append(res1.detach().cpu().numpy())
        res2 = model_LSTM(xb)
        m2.append(res2.detach().cpu().numpy())
        res3 = model_GRU(xb)
        m3.append(res3.detach().cpu().numpy())
m1 = np.concatenate(m1)
m2 = np.concatenate(m2)
m3 = np.concatenate(m3)
m1 = scaler.inverse_transform(m1)
m2 = scaler.inverse_transform(m2)
m3 = scaler.inverse_transform(m3)    
```

```python
real  = scaler.inverse_transform(y[tr_size:])
```
